```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_excel("golf_df.xlsx")
dados.head()
```

```python
x = pd.get_dummies(dados[["clima", "temp", "umidade", "vento"]])
x.head()
```

```python
# Jogar é a nossa variavel Target, e como  a árvore de decisão permite variavel categorica materemos assim.
y = dados[["jogar"]]
y
```

```python
from sklearn import tree
```

```python
arvore = tree.DecisionTreeClassifier()
modelo = arvore.fit(x,y)
```

```python
modelo
```

```python
fig = plt.figure(figsize=(70,30))

fig = tree.plot_tree(modelo,
                     feature_names=x.columns,
                     class_names=["yes","no"],
                     filled = True
                     )
```
