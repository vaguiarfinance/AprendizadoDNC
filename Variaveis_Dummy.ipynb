```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
limite = pd.read_excel("Limite_Credito.xlsx")
limite.head()
```

```python
# Já que o modelo não realiza a leitura variáveis categoricas, por isso vamos transformar para Dummys

limite = pd.get_dummies(limite, prefix=["Escolaridade"])
limite.head()
```

```python
# Vamos tirar a correlação entre as variáveis

limite.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
function = "LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos"
```

```python
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

```python
# Já que temos poucas variáveis, optamos por fazer o método Backward e iremos excluir a idade, uma vez que ficou em 91% (P-Value)

function = "LimitedoChequeEspecial~RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos"
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

```python
# Excluindo do rendimento total que ficou em 18% (P-Value) e -1 é para excluir a variavel Intercept

function = "LimitedoChequeEspecial~Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos -1"
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

```python
# Excluindo a Escolaridade_Superior_Pos, pois ficou em 93%
function = "LimitedoChequeEspecial~Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio -1"
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

Assim chegamos a 90% de acuracia, alegando que a variavel Limite do Cheque é especial é extremamente afetada pela oscilação do Salário, Limite de Crédito Imediato e Escolaridade (Fundamental/Médio)

```python
#Foram incluídas novas variáveis no Dataset

new_limit = pd.read_excel("Limite_Credito_Dummy.xlsx")
new_limit.head()
```

```python
new_limit = pd.get_dummies(data=new_limit, columns=["Escolaridade","Gênero","Região"], prefix=["Escolaridade","Gênero","Região"])
new_limit.head()
```

```python
new_limit.corr()
```

```python
function = "LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Gênero_RJ+Gênero_SP+Região_F+Região_M"
model = smf.ols(formula=function , data=new_limit).fit()
print(model.summary())
```

```python
function = "LimitedoChequeEspecial~RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Superior_Pos+Gênero_RJ+Gênero_SP+Região_F+Região_M"
model = smf.ols(formula=function , data=new_limit).fit()
print(model.summary())
```

```python
function = "LimitedoChequeEspecial~Salario+Gênero_RJ+Gênero_SP+Região_F+Região_M -1"
model = smf.ols(formula=function , data=new_limit).fit()
print(model.summary())
```

Com isso temos uma explicabilidade de 87%, sobre as variaveis explicativas { Salário | Genero (SP/RJ) | Região (F/M)} em função da variação da Target { Limite do Cheque Especial }
